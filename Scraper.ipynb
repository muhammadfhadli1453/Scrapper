```python
from selenium import webdriver
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary  
from selenium.common.exceptions import NoSuchElementException  
from selenium.webdriver.common.keys import Keys  
from bs4 import BeautifulSoup
import MySQLdb
import progressbar
import os
```

```python
db = MySQLdb.connect("localhost","root","", "company")
cur = db.cursor()
clear = lambda: os.system('cls') #on Windows System
```

```python
def get_html(url):
    binary = FirefoxBinary(r"C:\Program Files\Mozilla Firefox\firefox.exe")
    fp = webdriver.FirefoxProfile()
    browser = webdriver.Firefox(firefox_binary=binary, firefox_profile=fp)
    
    browser.get(url)  
    html_source = browser.page_source  
    browser.quit()
    soup = BeautifulSoup(html_source,'html.parser')
    
    return soup
```

```python
def get_company_data(job_html):
    data = []
    try:
        data.append(job_html.find('div', attrs={'class': 'logo_sm_wrap'}).find('img')['src'])
    except AttributeError:
#         data['company_logo'] = 'No Data'
        data.append("No Data")

    try:
        data.append(job_html.find('div', attrs={'class': 'company_name'}).find('a').text)
#         data['company_name'] = job_html.find('div', attrs={'class': 'company_name'}).find('a').text
    except:
#         data['company_name'] = 'No Data'
        data.append("No Data")


    try:
        data.append(job_html.find('p', attrs={'id': 'address'}).text)
#         data['company_location'] = job_html.find('p', attrs={'id': 'address'}).text
    except AttributeError:
#         data['company_location'] = 'No Data'
        data.append("No Data")


    try:
        data.append(job_html.find('h1', attrs={'id': 'position_title'}).text)
#         data['company_job_title'] = job_html.find('h1', attrs={'id': 'position_title'}).text
    except AttributeError:
#         data['company_job_title'] = "No Data"
        data.append("No Data")

        
    try:
        data.append(job_html.find('p', attrs={'id': 'company_industry'}).text)
#         data['company_industry'] = job_html.find('p', attrs={'id': 'company_industry'}).text
    except AttributeError:
#         data['company_industry'] = 'No Data'
        data.append("No Data")


    try:
        data.append(job_html.find('p', attrs={'id': 'company_size'}).text)
#         data['company_size'] = job_html.find('p', attrs={'id': 'company_size'}).text
    except AttributeError:
#          data['company_size'] = 'No Data'
        data.append("No Data")

        
    try:
        data.append(job_html.find('div', attrs={'class', 'unselectable wrap-text'}))
#         data['company_description'] = job_html.find('div', attrs={'class', 'unselectable wrap-text'})
    except AttributeError:
#         data['company_description'] = 'No Data'
        data.append("No Data")

        
    try:
        data.append(job_html.find('p', attrs={'id': 'posting_date'}).find('span').text)
#         data['company_date'] = job_html.find('p', attrs={'id': 'posting_date'}).find('span').text
    except AttributeError:
#         data['company_date'] = 'No Data'  
        data.append("No Data")

    
    return data
```

```python
# url = 'https://www.jobstreet.co.id/id/job-search/job-vacancy.php?ojs=6'

page = 1
multipage_url = 'https://www.jobstreet.co.id/id/job-search/job-vacancy/'
table = []
```

```python
while page < 3: # This mean we will only scrap page 1-2. If you want the Scrapper to scrap all the page, just change "page < 3" into "True"
    print("Accessing Page "+(str(page))+"...")
    url = multipage_url + str(page) + '/'
    soup = get_html(url)
    data = soup.findAll('div',attrs={'class':'position-title'})

    if len(data) == 0:
        break
    else:
        page += 1
        for div in progressbar.progressbar(data):
            links = div.findAll('a')
            for a in links:
                if ((a['href'].find('sectionRank') != -1)):
                    job_html = get_html(a['href'])
                    table.append(get_company_data(job_html))
        clear()
```
